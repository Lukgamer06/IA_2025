# K-Vecinos Segundo Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
```

## Datos

```python
data = pd.read_csv('reviews_sentiment.csv', sep=";")

# Filtrar datos v√°lidos
data = data.dropna(subset=["wordcount", "sentimentValue", "Star Rating"])
X = data[["wordcount", "sentimentValue"]].values
y = data["Star Rating"].astype(int).values

# Separar por clase
bueno = data[data["Star Rating"] >= 4]
malo = data[data["Star Rating"] <= 2]

# Graficar
plt.figure(figsize=(10, 6))
plt.scatter(bueno["wordcount"], bueno["sentimentValue"], marker="*", s=150, color="skyblue", label="Estrellas ‚â• 4 (Bueno)")
plt.scatter(malo["wordcount"], malo["sentimentValue"], marker="*", s=150, color="red", label="Estrellas ‚â§ 2 (Malo)")

plt.xlabel("Cantidad de palabras (wordcount)")
plt.ylabel("Valor de sentimiento (sentimentValue)")
plt.title("Distribuci√≥n de rese√±as por sentimiento y longitud")
plt.legend(bbox_to_anchor=(1, 0.3))
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Preparacion de los datos

```python
escalador = preprocessing.MinMaxScaler()
X_escalado = escalador.fit_transform(X)
```

## Creacion del modelo KNN

```python
# Nuevo dato
nuevo = escalador.transform([[30, 0.5]])

metricas = [
    ('euclidean', {}),
    ('manhattan', {}),
    ('minkowski', {'p': 3}),
    ('hamming', {})
]

n_neigbors =[ 1, 3, 5, 7, 10]

# Datos sint√©ticos para graficar
wordcounts = np.array([np.arange(1, 101)] * 100).reshape(-1, 1)
sentiments = np.array([np.linspace(-1.5, 2.5, 100)] * 100).reshape(-1, 1)
todos = pd.DataFrame(np.hstack((wordcounts, sentiments)), columns=["wordcount", "sentimentValue"])
solicitantes = escalador.transform(todos)

resultados = []
for nombre_metric, params in metricas:
    for k in n_neigbors:
        print(f"\nüîç Distancia: {nombre_metric} | Vecinos: {k}")

        modelo = KNeighborsClassifier(n_neighbors=k, metric=nombre_metric, **params)
        modelo.fit(X_escalado, y)

        # Predicci√≥n del nuevo dato
        pred = modelo.predict(nuevo)[0]
        probas = modelo.predict_proba(nuevo)[0]
        print(f"‚≠ê Predicci√≥n: {pred}")
        print(f"üìä Probabilidades: {probas}")

        # Evaluaci√≥n
        y_pred = modelo.predict(X_escalado)
        matriz = confusion_matrix(y, y_pred)
        acc = accuracy_score(y, y_pred)
        print("üìå Matriz de confusi√≥n:")
        print(matriz)
        print(f"‚úÖ Precisi√≥n: {acc:.4f}")

        resultados.append({
            "distancia": nombre_metric,
            "vecinos": k,
            "precision": acc
        })

        # Graficar resultados sint√©ticos
        todos["pred"] = modelo.predict(solicitantes)
        plt.figure(figsize=(8, 6))
        plt.title(f"Distancia: {nombre_metric} | Vecinos: {k}")
        for rating in sorted(todos["pred"].unique()):
            subset = todos[todos["pred"] == rating]
            plt.scatter(subset["wordcount"], subset["sentimentValue"], s=60, label=f"Predicci√≥n: {rating}", alpha=0.6)
        plt.scatter(30, 0.5, marker="P", s=250, color="black", label="Nuevo dato")
        plt.xlabel("Wordcount")
        plt.ylabel("Sentiment Value")
        plt.legend(bbox_to_anchor=(1, 0.3))
        plt.tight_layout()
        plt.show()

# Mejor combinaci√≥n
mejor = max(resultados, key=lambda x: x["precision"])
print("\nüèÜ Mejor combinaci√≥n:")
print(f"Distancia: {mejor['distancia']} | Vecinos: {mejor['vecinos']} | Precisi√≥n: {mejor['precision']:.4f}")
```
