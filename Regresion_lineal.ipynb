```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
```

# Salary

```python
data = pd.read_excel('Salary.xlsx')
data.head()
```

```python
df = pd.DataFrame(data) # Crear un DataFrame 
X = df[['YearsExperience']] # Caracter√≠stica independiente
Y = df['Salary'] # Variable dependiente

# Graficamos los datos
plt.scatter(X, Y, color='blue')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Experience vs Salary')
plt.show()
```

```python
# Ajustar los datos a un modelo de regresion lineal
model = LinearRegression()
model.fit(X,Y)

# Generar predicciones
Y_pred= model.predict(X)

# Hallar el R2
r2 = r2_score(Y,Y_pred)

print(r2)
```

```python
plt.scatter(X, Y, color='blue')
plt.plot(X, Y_pred, color='red', linewidth=5)
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Experience vs Salary')
plt.show()

print(f"Ecuacion de la recta: y = {model.coef_[0]:.2f}x + {model.intercept_:2f}")
print(f"Coeficiente de determinacion R^2: {r2:.2f}")
```

# Edad-Precio

```python
data2 = pd.read_csv('dataset_edad_precio_seguro.csv')
data2.head()
```

```python
X2 = data2['Edad']
Y2 = data2['PrecioSeguro']

# Graficamos los datos
plt.scatter(X2, Y2, color='blue')
plt.xlabel('Edad')
plt.ylabel('Precio del Seguro')
plt.title('Edad vs Precio del Seguro')
plt.show()
```

```python
# Ajustar los datos a un modelo de regresion lineal
model2 = LinearRegression()
model2.fit(X2.values.reshape(-1,1),Y2)


# Generar predicciones
Y2_pred= model2.predict(X2.values.reshape(-1,1))

# Hallar el R2
r2_ = r2_score(Y2,Y2_pred)

print(r2_)
```

```python
plt.scatter(X2, Y2, color='blue')
plt.plot(X2, Y2_pred, color='red', linewidth=5)
plt.xlabel('Edad')
plt.ylabel('Precio del Seguro')
plt.title('Edad vs Precio del Seguro')
plt.show()

print(f"Ecuacion de la recta: y = {model2.coef_[0]:.2f}x + {model2.intercept_:2f}")
print(f"Coeficiente de determinacion R^2: {r2_:.2f}")
```

# House Prices

```python
data3 = pd.read_csv('house_prices_large.csv')
data3.head()
```

```python
X3 = data3[['House_Size']]
Y3 = data3['Price']

# Graficamos los datos
plt.scatter(X3, Y3, color='blue')
plt.xlabel('House Size (sq ft)')
plt.ylabel('Price ($)')
plt.title('House Size vs Price')
plt.show()
```

```python
# Ajustar los datos a un modelo de regresion lineal
model3 = LinearRegression()
model3.fit(X3,Y3)

# Generar predicciones
Y3_pred= model3.predict(X3)

# Hallar el R2
r2_3 = r2_score(Y3,Y3_pred)
print(r2_3)
```

```python
plt.scatter(X3, Y3, color='blue')
plt.plot(X3, Y3_pred, color='red', linewidth=5)
plt.xlabel('House Size')
plt.ylabel('Price')
plt.title('House Size vs Price')
plt.show()
print(f"Ecuacion de la recta: y = {model3.coef_[0]:.2f}x + {model3.intercept_:2f}")
print(f"Coeficiente de determinacion R^2: {r2_3:.2f}")
```
