## Analizando la relacion entre las caracteristicas con la probabilidad de supervivencia


```python
df.columns
```

```python
def bar_chart(feature):
    survived = df[df['Survived'] == 1][feature].value_counts()
    dead = df[df['Survived'] == 0][feature].value_counts()
    test = pd.DataFrame([survived, dead])
    test.index = ['Survived', 'Dead']
    test.plot(kind='bar', stacked=True, figsize=(10, 5))
    plt.xlabel(feature)
    plt.ylabel('Number of passengers')
    plt.title('Survival count in relation to ' + feature)
    plt.show()
```

```python
bar_chart('Sex')
```

La Grafica Seala que las mujeres en el titanic tienen casi 3 veces de probabilidad que los hombres

```python
bar_chart('Pclass')
```

La Grafica seala que la clase alta tuvo mayor probabilidad de supervivencia con respecto a las **dems clases**. ademas de que la tercera clase fue la que mas fallecio en comparacion a las **dems clases**

```python
bar_chart('Embarked')
```

La grafica confirma que:
* Una persona que **aborda desde Cheboug** tuvo una ligera probabilidad de **sobrevivir**
* Si una persona que **aborda desde Southampton y Queenstown** probablemente **Murio**
